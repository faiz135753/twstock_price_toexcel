```python
import requests
from bs4 import BeautifulSoup
import time     #暫留時間
import random   #隨機停留
import openpyxl #串接excel
import datetime #引用產生日期


stock_total_list = []
pagex = 0
while pagex < 43:     
    pagex+=1
    res = requests.get(
    'https://histock.tw/stock/rank.aspx?&m=0&p='+str(pagex)+'&d=0'
    )
    soup = BeautifulSoup(res.text)  
    for i in range(0,24):
        try:
            stock_name = soup.find_all('table',class_="gvTB")[0].find_all('tr',class_="alt-row")[i].find_all('td')[0].text
        except IndexError:
            continue
        stock_total_list.append(stock_name)
        print(stock_name,pagex)
count=0    

for stock_range in stock_total_list:
    time_open='2020/01/01'
    time_close='2022/04/27'
    res_stock_data =requests.get("https://www.cnyes.com/twstock/ps_historyprice.aspx?code="+stock_range+"&ctl00$ContentPlaceHolder1$startText="+time_open+"&ctl00$ContentPlaceHolder1$endText="+time_close)
    soup_stock_data=BeautifulSoup(res_stock_data.text)
    wb = openpyxl.Workbook()
    ws_all_list=wb.active
    ws_all_list['a1'] = "日期"
    ws_all_list['b1'] = "開盤"
    ws_all_list['c1'] = "最高"
    ws_all_list['d1'] = "最低"
    ws_all_list['e1'] = "收盤"
    ws_all_list['f1'] = "漲跌"
    ws_all_list['g1'] = "漲%"
    ws_all_list['h1'] = "成交量"
    ws_all_list['i1'] = "成交金額"
    ws_all_list['j1'] = "本益比"
    wb.save("C:\\Users\\faiz6\\stock_data\\"+stock_range+'data.xlsx')
    ws_all_list.title = stock_range
    try:
        for data_stock in range(1,9999999999):
            data_stock_1=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[0].text)
            data_stock_2=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[1].text)
            if len(data_stock_2)>=8:
                data_stock_2=data_stock_2.replace(',', '')
                data_stock_2=float(data_stock_2)*1
            else:    
                data_stock_2=float(data_stock_2)*1
            data_stock_3=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[2].text)
            if len(data_stock_3)>=8:
                data_stock_3=data_stock_3.replace(',', '')
                data_stock_3=float(data_stock_3)*1
            else:    
                data_stock_3=float(data_stock_3)*1
            data_stock_4=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[3].text)
            if len(data_stock_4)>=8:
                data_stock_4=data_stock_4.replace(',', '')
                data_stock_4=float(data_stock_4)*1
            else:
                data_stock_4=float(data_stock_4)*1
            data_stock_5=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[4].text)
            if len(data_stock_5)>=8:
                data_stock_5=data_stock_5.replace(',', '')
                data_stock_5=float(data_stock_5)*1
            else:
                data_stock_5=float(data_stock_5)*1
            data_stock_6=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[5].text)
            data_stock_7=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[6].text)
            data_stock_8=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[7].text)
            data_stock_9=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[8].text)
            data_stock_10=(soup_stock_data.find_all('div',class_="tab")[0].find_all('tr')[data_stock].find_all('td')[9].text)
            ws_all_list.append([data_stock_1,data_stock_2,data_stock_3,data_stock_4,data_stock_5,data_stock_6,data_stock_7,data_stock_8,data_stock_9,data_stock_10])
            wb.save("C:\\Users\\faiz6\\stock_data\\"+stock_range+'data.xlsx')
    except:
        pass
```
